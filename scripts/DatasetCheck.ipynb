```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
dnd = []
```

```python
data = np.load('../data/20190702_20part_PtOrder_v1/Wlnu.npy')
print(data.shape)
```

```python
for i in range(data.shape[2]):
    x = data[:,0,i]
    f = plt.figure(i)
    r = (np.min(x), np.max(x))
    plt.hist(x, bins=20, range=r, alpha=0.4, density=True)
    plt.xlabel('Part 0 ')
    dnd.append(f)
    plt.yscale('log')
    print(np.min(x))
```
